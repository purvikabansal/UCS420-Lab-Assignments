```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed using roll number
np.random.seed(102317121)

# Generate random sales data for 12 months and 4 product categories
sales_data = np.random.randint(1000, 5000, (12, 4))

# Define month names and product categories
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

# Create a DataFrame to store the sales data
df = pd.DataFrame(sales_data, columns=categories, index=months)

# Display first few rows of the DataFrame
print(df.head())
```

```python
# Display summary statistics of the sales data
print(df.describe())

# Calculate total sales for each category
total_sales_per_category = df.sum()
print(total_sales_per_category)

# Add a column for total sales per month
df['Total Sales'] = df.sum(axis=1)
print(df['Total Sales'])

# Calculate average sales growth between consecutive months for each category
growth_per_category = df.pct_change().mean()
print("Average Sales Growth per Category", growth_per_category)

# Calculate the month-to-month growth rate for total sales
df['Growth Rate'] = df['Total Sales'].pct_change() * 100
print("\nGrowth Rate:\n", df['Growth Rate'])
```

```python
# Apply discount based on roll number (even: 10% on Electronics, odd: 15% on Clothing)
df['Electronics'] *= 0.9  # 10% discount on Electronics
print("\nDiscounted Electronics Sales:\n", df['Electronics'])

# Plot sales trends for each category
plt.figure(figsize=(10, 6))
for category in categories:
    plt.plot(df.index, df[category], marker='o', label=category)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trends")
plt.legend()
plt.show()

# Create box plots for sales distribution
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[categories])
plt.title("Sales Distribution per Category")
plt.show()
```

```python
# Q2: Given array
array = np.array([[1, -2, 3], [-4, 5, -6]])

# Element-wise absolute value
abs_array = np.abs(array)
print("Absolute Values", abs_array)

# Percentiles of the flattened array
percentiles_flattened = np.percentile(array, [25, 50, 75])
print(" Percentiles (Flattened)", percentiles_flattened)

# Percentiles for each column
percentiles_columns = np.percentile(array, [25, 50, 75], axis=0)
print("Percentiles (Columns)", percentiles_columns)

# Percentiles for each row
percentiles_rows = np.percentile(array, [25, 50, 75], axis=1)
print(" Percentiles (Rows)\n", percentiles_rows)

# Mean, Median, Standard Deviation of the flattened array
mean_flattened = np.mean(array)
median_flattened = np.median(array)
std_flattened = np.std(array)
print("\nMean, Median, Standard Deviation (Flattened)", mean_flattened, median_flattened, std_flattened)

# Mean, Median, Standard Deviation for each column
mean_columns = np.mean(array, axis=0)
median_columns = np.median(array, axis=0)
std_columns = np.std(array, axis=0)
print("Mean, Median, Standard Deviation (Columns):\n", mean_columns, median_columns, std_columns)

# Mean, Median, Standard Deviation for each row
mean_rows = np.mean(array, axis=1)
median_rows = np.median(array, axis=1)
std_rows = np.std(array, axis=1)
print("\nMean, Median, Standard Deviation (Rows):\n", mean_rows, median_rows, std_rows)
```

```python
# Q3: Given array
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

# Compute floor values
floor_values = np.floor(a)
print("Floor Values:\n", floor_values)

# Compute ceiling values
ceiling_values = np.ceil(a)
print("Ceiling Values:\n", ceiling_values)

# Compute truncated values
truncated_values = np.trunc(a)
print("Truncated Values:\n", truncated_values)

# Compute rounded values
rounded_values = np.round(a)
print("Rounded Values:\n", rounded_values)
```

```python
# Q4: Swap two elements in a list using a temporary variable

def swap_list_elements(lst, idx1, idx2):
    temp = lst[idx1]
    lst[idx1] = lst[idx2]
    lst[idx2] = temp
    return lst

# Example list
my_list = [10, 20, 30, 40, 50]

# Swap elements at index 1 and 3
swapped_list = swap_list_elements(my_list, 1, 3)
print("List after swapping:", swapped_list)
```

```python
# Q5: Swap two elements in a set by converting it to a list first

def swap_set_elements(s, val1, val2):
    lst = list(s)
    if val1 in lst and val2 in lst:
        idx1, idx2 = lst.index(val1), lst.index(val2)
        lst[idx1], lst[idx2] = lst[idx2], lst[idx1]
        return set(lst)
    return s

# Example set
my_set = {10, 20, 30, 40, 50}

# Swap elements 20 and 40
swapped_set = swap_set_elements(my_set, 20, 40)
print("Set after swapping:", swapped_set)
```
