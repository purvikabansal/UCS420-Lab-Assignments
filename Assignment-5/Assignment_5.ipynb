```python
import numpy as np
import matplotlib.pyplot as plt

# Q1
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print("Total Sum:", np.sum(gfg))
print("Row-wise Sum:", np.sum(gfg, axis=1))
print("Column-wise Sum:", np.sum(gfg, axis=0))
```

```python
import numpy as np

# Q2(a)
arr1 = np.array([10, 52, 62, 16, 16, 54, 453])
sorted_arr = np.sort(arr1)
sorted_indices = np.argsort(arr1)
smallest_4 = sorted_arr[:4]
largest_5 = sorted_arr[-5:]

print("Sorted Array:", sorted_arr)
print("Indices of Sorted Array:", sorted_indices)
print("4 Smallest Elements:", smallest_4)
print("5 Largest Elements:", largest_5)

# Q2(b)
arr2 = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
int_elements = arr2[arr2 == arr2.astype(int)]
float_elements = arr2[arr2 != arr2.astype(int)]

print("Integer Elements:", int_elements)
print("Float Elements:", float_elements)
```

```python
# Q3
X_val = ord('P') + ord('B')
sales_data = np.array([X_val, X_val + 50, X_val + 100, X_val + 150, X_val + 200])
print("dataset:", sales_data)

tax_rate = ((X_val % 5) + 5) / 100
taxed_sales = sales_data * (1 + tax_rate)
print(taxed_sales)

discounted_sales = []
for sale in sales_data:
    if sale < X_val + 100:
        discounted_sales.append(sale * 0.95)
    else:
        discounted_sales.append(sale * 0.90)
print(discounted_sales)

num_weeks = 3
weekly_sales = np.vstack([sales_data] * num_weeks)
growth_factor = 1.02 ** np.arange(num_weeks)[:, None]
weekly_sales_adjusted = weekly_sales * growth_factor

print(weekly_sales_adjusted)
```

```python
#Q4
x = np.linspace(-10, 10, 100)
y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)

plt.plot(x, y1, label="y = x^2", color='red')
plt.title("y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

plt.plot(x, y2, label="y = sin(x)", color='green')
plt.title("y = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
plt.plot(x, y3, label="y = e^x", color='yellow')
plt.title("y = e^x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.show()
```

```python
plt.plot(x, y4, label="y = log(|x| + 1)", color='red')
plt.title("y = log(|x| + 1)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.show()
```
