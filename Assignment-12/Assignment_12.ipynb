```python
# Step 1: Install ChatterBot and TextBlob
!git clone https://github.com/gunthercox/ChatterBot.git
%cd ChatterBot
!pip install -e .

!pip install textblob

# Step 2: Import Libraries
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer
from textblob import TextBlob

# Step 3: Create the HealthBot
health_bot = ChatBot(
    "HealthBot",
    logic_adapters=[
        "chatterbot.logic.BestMatch"
    ]
)

# Step 4: Training Data
trainer = ListTrainer(health_bot)

training_data = [
    "Hi",
    "Hello! I am HealthBot. How can I assist you today?",
    "I have a headache",
    "You should rest, stay hydrated, and take a mild pain reliever if needed.",
    "What should I eat for a cold?",
    "Warm fluids, soups, citrus fruits, and light meals help during a cold.",
    "I feel stressed",
    "Try relaxation techniques like deep breathing, meditation, or talking to someone you trust.",
    "Bye",
    "Goodbye! Stay healthy."
]

# Step 5: Train the chatbot
trainer.train(training_data)

# Step 6: Create a Chat Loop with Sentiment Analysis
print("Start talking with HealthBot! (type 'exit' to stop)")

while True:
    user_input = input("You: ")
    if user_input.lower() == 'exit':
        print("HealthBot: Take care! Goodbye.")
        break

    # Sentiment Analysis
    blob = TextBlob(user_input)
    sentiment = blob.sentiment.polarity

    if sentiment < -0.3:
        print("HealthBot: I sense you're feeling upset. It's important to take care of your mental health. ðŸ’™")
    elif sentiment > 0.5:
        print("HealthBot: You seem happy today! That's great to hear! ðŸŒŸ")

    # ChatBot Response
    response = health_bot.get_response(user_input)
    print(f"HealthBot: {response}")
```
